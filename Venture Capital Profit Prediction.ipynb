# Multiple Linear Regression

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
ds = pd.read_csv('50_Startups.csv')
```

```python
ds.head()
```

```python
X = ds.iloc[:,:-1].values
y = ds.iloc[:, -1].values
```

```python
print(X.shape)
print(y.shape)
```

## There is no missing data, so we take no action for this step.

## Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le_X = LabelEncoder()
ohe_X = OneHotEncoder(categorical_features = [-1])

X[:,-1] = le_X.fit_transform(X[:, -1])
X = ohe_X.fit_transform(X).toarray()
```

```python
print(X)
```

## Avoiding the Dummy Variable Trap

```python
# To keep track of the categorical variables:
# (1,0,0) -> California, (0,1,0) -> Florida, (0,0,1) -> New York
# After this step:
# (0,0) -> California, (1,0) -> Florida, (0,1) -> New York

X = X[:,1:]
print(X.shape)
```

## Splitting the Dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## We do not require Feature Scaling as the linear_model library takes care of this for us.

## Fitting the Linear Regression to the dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Predicting the Test Test Values

```python
y_pred = lr.predict(X_test)
```

```python
print(y_test, y_pred)
```

## Error Metrics (Mean Absolute Error, Mean Squared Error and Root Mean Squared Error)

```python
from sklearn import metrics
ABS = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

print(ABS)
print(MSE)
print(RMSE)
```

## Making a single prediction (for example the profit of a startup with R&D Spend = 160000, Administration Spend = 130000, Marketing Spend = 300000 and State = 'California')

```python
lr.predict([[0, 0, 160000, 130000, 300000]])
```

## Backwards Elimination

```python
import statsmodels.api as sm
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)
X_opt = X[:, [0,1,2,3,4,5]]
print(X_opt)
Reg_OLS = sm.OLS(endog = y, exog = X_opt).fit()
Reg_OLS.summary()
```

```python
X_opt = X[:, [0,1,3,4,5]]
Reg_OLS = sm.OLS(endog = y, exog = X_opt).fit()
Reg_OLS.summary()
```

```python
X_opt = X[:, [0,3,4,5]]
Reg_OLS = sm.OLS(endog = y, exog = X_opt).fit()
Reg_OLS.summary()
```

```python
# From the below we observe that regressor takes the equation: 
# Profit = 46980 + 0.7966*(R&D SPend) + 0.0299*(Marketing Spend)

X_opt = X[:, [0,3,5]]
Reg_OLS = sm.OLS(endog = y, exog = X_opt).fit()
Reg_OLS.summary()
```
